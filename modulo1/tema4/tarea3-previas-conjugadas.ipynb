## 1. Beta es previa conjugada de Bernoulli

Completar la siguiente tabla de actualización:

| Hipótesis | Datos | Previa                                         | Verosimilitud               | Posterior                                               |
| --------- | ----- | ---------------------------------------------- | --------------------------- | ------------------------------------------------------- |
| $\theta$  | $x$   | $\text{Beta}(a, b)$                            | $\text{Bernoulli}(\theta)$  | -                                                       |
| $\theta$  | $x=1$ | $p(\theta)=c_1\theta^{a-1} (1 - \theta)^{b-1}$ | $p(x\|\theta)=\theta$       | ?                                                      |
| $\theta$  | $x=0$ | $p(\theta)=c_1\theta^{a-1} (1 - \theta)^{b-1}$ | $p(x\|\theta)=(1 - \theta)$ | ?                                                    |

Si se tienen constantes en la tabla, como por ejemplo $c_1$, especificarlas.


$ x = 1$:

La verosimilitud es $ p(x \mid \theta) = \theta $.

La posterior no normalizada se obtiene multiplicando la previa y la verosimilitud:

$$
p(\theta \mid x = 1) \propto p(\theta) \cdot p(x \mid \theta) = \left( c_1 \theta^{a-1} (1 - \theta)^{b-1} \right) \cdot \theta = c_1 \theta^a (1 - \theta)^{b-1}
$$

Entonces:

$$
p(\theta \mid x = 1) = \text{Beta}(a+1, b)
$$


Para $x = 0 $:

La verosimilitud es $( p(x \mid \theta) = 1 - \theta )$.

$$
p(\theta \mid x = 0) \propto p(\theta) \cdot p(x \mid \theta) = \left( c_1 \theta^{a-1} (1 - \theta)^{b-1} \right) \cdot (1 - \theta) = c_1 \theta^{a-1} (1 - \theta)^b
$$

Entonces:

$$
p(\theta \mid x = 0) = \text{Beta}(a, b+1)
$$






## 2. Beta es previa conjugada de geométrica

Completar la siguiente tabla de actualización:

| Hipótesis | Datos | Previa                                         | Verosimilitud                     | Posterior                                               |
| --------- | ----- | ---------------------------------------------- | --------------------------------- | ------------------------------------------------------- |
| $\theta$  | $x$   | $\text{Beta}(a, b)$                            | $\text{Geometrica}(\theta)$       | ?                                                       |
| $\theta$  | $x$   | $p(\theta)=c_1\theta^{a-1} (1 - \theta)^{b-1}$ | $p(x\|\theta)=\theta(1-\theta)^x$ | $\text{Beta}(a + 1, b + x)$                                                       |

Si se tienen constantes en la tabla, como por ejemplo $c_1$, especificarlas completamente.

La previa es $ \text{Beta}(a, b) $, es decir:

$$
p(\theta) = c_1 \theta^{a - 1} (1 - \theta)^{b - 1}
$$

donde

$$
c_1 = \frac{\Gamma(a + b)}{\Gamma(a) \Gamma(b)}
$$

es una constante.

La verosimilitud de una distribución geométrica, esta dada por:

$$
p(x | \theta) = \theta (1 - \theta)^x
$$

Multiplicamos la previa y la verosimilitud:

$$
p(\theta | x) \propto p(\theta) \cdot p(x | \theta)
$$

Sustituyendo:

$$
p(\theta | x) \propto \left( \theta^{a - 1} (1 - \theta)^{b - 1} \right) \cdot \left( \theta (1 - \theta)^x \right)
$$

Agrupando los términos:

$$
p(\theta | x) \propto \theta^{a - 1 + 1} (1 - \theta)^{b - 1 + x}
$$

Entonces, la posterior es una distribución Beta con los siguientes parámetros actualizados:

$$
\text{Beta}(a + 1, b + x)
$$


## 3. Gamma es previa conjugada de Poisson

Supongamos que tenemos un dato $x \in \mathbb{N}$ que tiene una verosimilitud de $\text{Poisson}(\theta)$:

$$
p(x | \theta) = \frac{\theta^x e^{-\theta}}{x!}.
$$

Ahora, si elegimos la distribución previa sobre el parámetro $\theta$ como una distribución $\text{Gamma}(a, b)$:

$$
p(\theta) = \frac{b^a}{\Gamma(a)} \theta^{a - 1} \exp\{-b \theta\}
$$

Encontrar completamente la densidad de probabilidad posterior

$$
p(\theta | x)
$$

##### Verosimilitud (distribución de Poisson):

$$
p(x | \theta) = \frac{\theta^x e^{-\theta}}{x!}
$$

##### Distribución previa (Gamma):

$$
p(\theta) = \frac{b^a}{\Gamma(a)} \theta^{a - 1} e^{-b \theta}
$$

##### Multiplicamos la verosimilitud y la previa para obtener la posterior no normalizada:

$$
p(\theta | x) \propto p(\theta) \cdot p(x | \theta) = \left( \frac{b^a}{\Gamma(a)} \theta^{a - 1} e^{-b \theta} \right) \cdot \left( \frac{\theta^x e^{-\theta}}{x!} \right)
$$

##### Simplificamos agrupando términos en $\theta$ y constantes:

$$
p(\theta | x) \propto \theta^{a - 1 + x} e^{-(b + 1) \theta}
$$


$ p(\theta | x) = \text{Gamma}(a + x, b + 1) $


## 4. Fórmulas de actualización normal

Supongamos que tenemos un dato $x \sim \text{Normal}(\theta, \sigma^2)$, donde la varianza $\sigma^2$ es conocida. Es decir, la media $\theta$ es nuestro parámetro deconocido de interés. Si elegimos una previa normal

$$
\theta \sim \text{Normal}(\mu_0, \sigma_0^2),
$$

entonces la posterior también es normal $(\theta | x) \sim \text{Normal}(\mu_1, \sigma_1^2)$. Demostrar que

$$
\frac{\mu_1}{\sigma_1^2} = \frac{\mu_0}{\sigma_1^2} + \frac{x}{\sigma^2}, \qquad \frac{1}{\sigma_1^2} = \frac{1}{\sigma_0^2} + \frac{1}{\sigma^2},
$$

o equivalentemente

$$
\mu_1 = \frac{\mu_0 / \sigma_0^2 + x / \sigma^2}{1 / \sigma_0^2 + 1 / \sigma^2} = \frac{\sigma^2 \mu_0 + \sigma_0^2 x}{\sigma_0^2 + \sigma^2}, \qquad \sigma_1^2 = \frac{1}{1 / \sigma_0^2 + 1 / \sigma^2} = \frac{\sigma_0^2\sigma^2}{\sigma_0^2 + \sigma^2}
$$

##### Distribuciones Previas y Verosimilitud:

La verosimilitud del dato $x$ dado el parámetro $\theta$ es:

$$
p(x | \theta) = \frac{1}{\sqrt{2\pi \sigma^2}} e^{-\frac{(x - \theta)^2}{2\sigma^2}}
$$

La distribución previa de $\theta$ es:

$$
p(\theta) = \frac{1}{\sqrt{2\pi \sigma_0^2}} e^{-\frac{(\theta - \mu_0)^2}{2\sigma_0^2}}
$$

##### Aplicación del Teorema de Bayes:

La posterior se puede calcular usando el teorema de Bayes:

$$
p(\theta | x) \propto p(x | \theta) p(\theta)
$$

Sustituyendo las expresiones:

$$
p(\theta | x) \propto e^{-\frac{(x - \theta)^2}{2\sigma^2}} e^{-\frac{(\theta - \mu_0)^2}{2\sigma_0^2}}
$$

Al multiplicar los exponentes:

$$
p(\theta | x) \propto e^{-\left( \frac{(x - \theta)^2}{2\sigma^2} + \frac{(\theta - \mu_0)^2}{2\sigma_0^2} \right)}
$$

##### Simplificación del Exponente:

Expandimos los términos:

$$
-\frac{(x - \theta)^2}{2\sigma^2} - \frac{(\theta - \mu_0)^2}{2\sigma_0^2} = -\left( \frac{\theta^2 - 2x\theta + x^2}{2\sigma^2} + \frac{\theta^2 - 2\mu_0\theta + \mu_0^2}{2\sigma_0^2} \right)
$$

Agrupamos:

$$
-\frac{1}{2} \left( \left(\frac{1}{\sigma^2} + \frac{1}{\sigma_0^2}\right) \theta^2 - \left(\frac{2x}{\sigma^2} + \frac{2\mu_0}{\sigma_0^2}\right) \theta + \left(\frac{x^2}{\sigma^2} + \frac{\mu_0^2}{\sigma_0^2}\right) \right)
$$

##### Forma Canónica de la Normal:

La forma general de la distribución normal es:

$$
p(\theta | x) \propto e^{-\frac{(\theta - \mu_1)^2}{2\sigma_1^2}}
$$

Debemos encontrar $\mu_1$ y $\sigma_1^2$ de manera que ambos exponentes coincidan.

Para que los coeficientes coincidan:

$$
\frac{1}{\sigma_1^2} = \frac{1}{\sigma_0^2} + \frac{1}{\sigma^2}
$$

Para que los coeficientes de $\theta$ coincidan:

$$
\frac{\mu_1}{\sigma_1^2} = \frac{\mu_0}{\sigma_0^2} + \frac{x}{\sigma^2}
$$

##### Resolviendo para \(\mu_1\) y \(\sigma_1^2\):

Simplificamos a:

$$
\sigma_1^2 = \frac{1}{\frac{1}{\sigma_0^2} + \frac{1}{\sigma^2}} = \frac{\sigma_0^2 \sigma^2}{\sigma_0^2 + \sigma^2}
$$

Para $\mu_1$:

$$
\mu_1 = \sigma_1^2 \left( \frac{\mu_0}{\sigma_0^2} + \frac{x}{\sigma^2} \right)
$$

Sustituyendo $\sigma_1^2$:

$$
\mu_1 = \frac{\sigma_0^2 \sigma^2}{\sigma_0^2 + \sigma^2} \left( \frac{\mu_0}{\sigma_0^2} + \frac{x}{\sigma^2} \right)
$$
$$
\mu_1 = \frac{\sigma^2 \mu_0 + \sigma_0^2 x}{\sigma_0^2 + \sigma^2}
$$

Por lo tanto:

$$
\frac{\mu_1}{\sigma_1^2} = \frac{\mu_0}{\sigma_0^2} + \frac{x}{\sigma^2}, \quad \frac{1}{\sigma_1^2} = \frac{1}{\sigma_0^2} + \frac{1}{\sigma^2}
$$
y
$$
\mu_1 = \frac{\sigma^2 \mu_0 + \sigma_0^2 x}{\sigma_0^2 + \sigma^2}, \quad \sigma_1^2 = \frac{\sigma_0^2 \sigma^2}{\sigma_0^2 + \sigma^2}
$$
